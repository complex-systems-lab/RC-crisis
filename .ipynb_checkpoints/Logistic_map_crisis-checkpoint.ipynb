Logistic map training with directional fibers

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import math
```

```python
class ReservoirComputer:
    
    def __init__(self, dim_system, dim_reservoir, rho, sigma):
        
        self.dim_system = dim_system
        self.dim_reservoir = dim_reservoir
        self.r_state = np.zeros(dim_reservoir)                                   
        self.A = generate_adjacency_matrix(dim_reservoir, rho, sigma)

        
        self.W_in = W_inn(dim_reservoir, dim_system)
        self.W_b = W_bb(dim_reservoir)
        self.W_out = np.zeros((dim_system, dim_reservoir))
        
        

    
    def advance_r_state(self, eps, u):
        alpha = 0.86
        self.r_state = ((1- alpha)* self.r_state) + alpha* np.tanh(np.matmul(self.A, self.r_state) + np.matmul(self.W_in, u.T) + self.W_b * (eps - 0.3))             # r_state is updated and stored as r(t+ delta t)
                                                                                    
        return self.r_state

    
    def train(self, trajectory_total, trajectory_new):                                                # Trajectory is the training data (initialized once model is called)
        
    
        R_total = np.empty((self.dim_reservoir, trajectory_total.shape[0]))
        eps_train = [3.92, 3.93, 3.94, 3.95]

        #eps_train = [3.92]

        div = int((trajectory_total.shape[0])/len(eps_train))
        
        for m in range(len(eps_train)):

            self.r_state = np.zeros(self.dim_reservoir)
            
            for i in range(div*m , div + div*m):
                
                R_total[:, i] = self.r_state
                u = trajectory_total[i]                                                   # stores i th row by default
                self.advance_r_state(eps_train[m], u)

        range_removal = [(0,29),(1000,1029),(2000, 2029),(3000, 3029)]
        #range_removal = [(0,29)]
        column_removal = []
        for start, end in range_removal:
            column_removal.extend(range(start, end + 1))
        
        # Remove the specified columns
        R_total = np.delete(R_total, column_removal, axis=1)  #removing the r_state transient (first 10 elements)         
        
        odd_column_indices = np.arange(R_total.shape[1])[1::2]    #squaring the odd rows                                                                   

        R_total[:, odd_column_indices] **= 2
        
        self.W_out = linear_regression(R_total, trajectory_new)

        #print(self.W_out)                  #To check the trained weights

    def v(self):
        return np.transpose(np.matmul(self.W_out, self.r_state))                   # Gives the predicted output


    def predict(self, eps_new, steps, test_trajectory, train_trajectory):
        
        prediction = np.zeros((steps, self.dim_system))
        prediction[0] = test_trajectory[0,:]

        warmup = train_trajectory[0:30, :]                       # warmup of reservoir

        self.r_state = np.zeros(self.dim_reservoir)

        for l in range(30):

            self.advance_r_state(eps_new, warmup[l])

        self.r_warmup = self.r_state
        
        for i in range(steps-1):
            
            self.advance_r_state(eps_new, prediction[i])
            v = self.v()
            prediction[i+1] = v
            
        return prediction

    def get_weights(self):
        return self.A, self.W_in, self.W_out, self.W_b, self.r_warmup
```

```python
def W_bb(dim_reservoir):

    np.random.seed(36)
    #W_b = (2* (np.random.rand(dim_reservoir)) - 0.5)* 1.15

    W_b = (2 * np.random.rand(dim_reservoir) - 1)* 1.15
    return W_b


#def W_inn(dim_reservoir, dim_system):

    #np.random.seed(85)
    #W_in = np.zeros((dim_reservoir, dim_system))
    #for i in range(dim_reservoir):
        
        #W_in[i, math.floor(i*(dim_system/dim_reservoir))] = (2*np.random.random()-1)*2.13

    #return W_in

def W_inn(dim_reservoir, dim_system):

    np.random.seed(85)
    W_in = 2 * 2.13 * (np.random.rand(dim_reservoir, dim_system) - 0.5)
    return W_in

def generate_adjacency_matrix(dim_reservoir, rho, sigma):
    
    graph = nx.gnp_random_graph(dim_reservoir, sigma, seed = 47)                           # Generates an Erdos-Renyi graph of nodes = dim_reservoir and probab of node connection = sigma
    
    np.random.seed(27)
    nx.set_edge_attributes(graph, {e: {'weight': np.random.uniform(0, 1)} for e in graph.edges})    
    graph = nx.to_numpy_array(graph)

    # Ensure random_array is of the same shape as the graph adjacency matrix.
    #np.random.seed(27)
    #random_array = 2 * sigma* (np.random.rand(dim_reservoir, dim_reservoir) - 0.5)

    # Multiply graph adjacency matrix with random values.
    #rescaled = graph * random_array
    rescaled = graph
    return scale_matrix(rescaled, rho)


def scale_matrix(A, rho):
    eigenvalues, _ = np.linalg.eig(A)
    max_eigenvalue = np.amax(eigenvalues)
    A = (A/np.absolute(max_eigenvalue)) * rho                 #this line makes the spectral radius of matrix A as rho
    return A
```

```python
def linear_regression(R, trajectory, beta=0.000001):
    Rt = np.transpose(R)
    inverse_part = np.linalg.inv(np.matmul(R, Rt) + beta * np.identity(R.shape[0]))                # Read Tikhonov Regularization
    return np.matmul(np.matmul(trajectory.T, Rt), inverse_part)                           # This is giving me the transpose of weight matrix basically with a shape (3,300)
```

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import dfibers.solvers as sv
import dfibers.fixed_points as fx
import dfibers.traversal as tv
import dfibers.numerical_utilities as nu
import dfibers.logging_utilities as lu

# Load model weights
A, W_in, W_out, W_b, rstate_initial = model.get_weights()

Lambda = A + np.matmul(W_in, W_out)
Omega = W_b
alpha = 0.86
N = Lambda.shape[0]

# Fixed-point detection grid (coarse steps)
bif_params_fp = np.arange(3.0, 4.1, 0.025)                                                       #np.arange(2.95, 4.15, 0.05)

# Bifurcation plot grid (fine steps)
bif_params_plot = np.arange(3.0, 4.1, 0.001)

# Storage for fixed points across bifurcation parameters
all_bif_fixed_points = []
all_bif_fixed_points_proj = []

# Function definitions
def f(V, bif_param):
    return -V + (1 - alpha) * V + alpha * np.tanh(Lambda @ V + Omega[:, None] * (bif_param - 0.3))
    
def Df(V, bif_param):
    N, P = V.shape
    gamma = Lambda @ V + Omega[:, None]*(bif_param - 0.3)
    sech2 = 1 - np.tanh(gamma)**2
    DfV = np.zeros((P, N, N))
    for p in range(P):
        D = np.diag(sech2[:, p])
        DfV[p,:,:] = alpha * (D @ Lambda - np.eye(N))
    return DfV



def ef(V):
    N, P = V.shape
    return 1e-10 * np.ones((N, P))

terminate = lambda trace: np.any(np.abs(trace.x[:N,:]) > 100)
compute_step_amount = lambda trace: (0.001, None, False)

# Duplicates check
duplicates = lambda V, v: (np.abs(V - v) < 1e-5).all(axis=0)

# Holds previous bif_param fixed points
fxpts_prev = []

# Loop over coarse bifurcation parameters
for bif_param in bif_params_fp:
    
    print(f"\nFinding fixed points for bif_param = {bif_param}")

    num_trials = 15
    all_fxpts_this_param = []

    for trial in range(num_trials):

        # Determine initial guess
        if fxpts_prev and trial < len(fxpts_prev):                                                   #if fxpts_prev < len() returns false if list empty, this is way of writing if the list is non empty and trial<len() holds true
            # Use fixed point from previous bif_param run, plus slight perturbation
            v0 = fxpts_prev[trial] + 0.001 * np.random.randn(N, 1)
            
        else:
            v0 = np.random.uniform(-1, 1, size=(N,1))
            

        # Random fiber direction
        c = np.random.randn(N,1)
        c /= np.linalg.norm(c)

        solution = sv.fiber_solver(
            f = lambda V: f(V, bif_param),
            ef = ef,
            Df = lambda V: Df(V, bif_param),
            v = v0,
            c = c,
            terminate = terminate,
            compute_step_amount = compute_step_amount,
            max_traverse_steps = 2000,
            max_solve_iterations = 32,
            logger = None,
        )

        fxpts = solution["Fixed points"]
        
        if fxpts.shape[1] > 0:
            # Sanitize the fixed points from this trial
            sanitized = fx.sanitize_points(
                fxpts,
                f = lambda V: f(V, bif_param),
                ef = ef,
                Df = lambda V: Df(V, bif_param),
                duplicates = duplicates
            )
            if sanitized.shape[1] > 0:
                all_fxpts_this_param.append(sanitized)
            else:
                print(f"  Trial {trial}: all fixed points eliminated during sanitization.")                  #in the above step, duplicates are removed and points with high residue are also removed
        else:
            print(f"  Trial {trial}: No fixed points found.")

    # Combine sanitized points from all trials
    if all_fxpts_this_param:
        fxpts_all = np.hstack(all_fxpts_this_param)

        # Final deduplication
        fxpts_all = fx.sanitize_points(
            fxpts_all,
            f = lambda V: f(V, bif_param),
            ef = ef,
            Df = lambda V: Df(V, bif_param),
            duplicates = duplicates
        )
        

        if fxpts_all.shape[1] > 0:
            print(f"Found {fxpts_all.shape[1]} unique fixed points at bif_param={bif_param}")
        
            stabilities = []  # True = stable, False = unstable
        
            # Print residuals and eigenvalues for stability analysis
            for i in range(fxpts_all.shape[1]):
                v_fp = fxpts_all[:, [i]]
                
                # Residual
                residual = np.linalg.norm(f(v_fp, bif_param))
                print(f"  FP {i}: residual = {residual:.3e}")
                
                # Jacobian at this fixed point
                J = Df(v_fp, bif_param)[0,:,:] + np.eye(v_fp.shape[0])                    #correct jacobian form
                
                # Eigenvalues
                eigvals = np.linalg.eigvals(J)
                max_abs = np.max(np.abs(eigvals))
                print(f"      max |eigenvalue| = {max_abs:.6f}")
        
                stabilities.append(max_abs <= 1.0)  # Stable if all eigenvalues <= 1 in magnitude
        
            # Project fixed points to output space
    
  

            projections = W_out @ fxpts_all
            all_bif_fixed_points.append(fxpts_all)
            all_bif_fixed_points_proj.append((bif_param, projections, stabilities))
        
            print(projections)
            
            # Save these fixed points as seeds for next bif_param
            fxpts_prev = [fxpts_all[:, [i]] for i in range(fxpts_all.shape[1])]

        else:
            print(f"No unique fixed points survived final sanitization at bif_param={bif_param}")
            fxpts_prev = []
    else:
        print(f"No fixed points found at bif_param={bif_param}")
        fxpts_prev = []


```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams.update({
    'font.size': 30,       # Increase font size for labels
    'font.family': 'sans-serif',
    'font.sans-serif': 'Helvetica Neue',
    'font.weight': 'normal',
    'font.stretch': 'condensed',    
    'axes.titlesize': 21,       # Title size
    'axes.labelweight': 'bold', # Bold axis labels
    'axes.titleweight': 'bold',  # Bold title
    'legend.fontsize': 25, # Increase font size for legend
    'axes.labelsize': 40,  # Increase font size for axis labels
    'xtick.labelsize': 44, # Increase font size for x-axis tick labels
    'ytick.labelsize': 44, # Increase font size for y-axis tick labels
    'lines.markersize': 10, # Increase marker size
    'text.usetex': True 
})

mu_coarse = np.arange(2.5, 3.5, 0.008)   # coarse step
mu_fine   = np.arange(3.5, 4.0, 0.0008)  # fine step
mu = np.concatenate([mu_coarse, mu_fine])  # merge

nt = 500
broadcast_shapee = 20

# Prepare mu broadcasted
mu_repeated = np.repeat(mu, broadcast_shapee)
mu_broadcasted = mu_repeated.reshape(len(mu), broadcast_shapee)

# Compute logistic-like map points
x_list1 = np.zeros((len(mu), broadcast_shapee))
for j in range(len(mu)):
    x = 0.7
    for _ in range(nt):  # Burn-in
        x = mu[j] * x - mu[j] * x**2
    for k in range(broadcast_shapee):
        x = mu[j] * x - mu[j] * x**2
        x_list1[j, k] = x

# Figure setup
plt.figure(figsize=(12, 8))

# Plot bifurcation diagram
for j in range(len(mu)):
    plt.plot(mu_broadcasted[j, :], x_list1[j, :],
             color='red', ls='', marker=',', alpha=0.6)

# Dashed lines: x = 0 and x = 1 - 1/mu (valid for mu > 1)
#plt.axhline(0, color='blue', ls='--', lw=1.5, label=r"$x = 0$")

valid_mu = mu[mu > 1]
plt.plot(valid_mu, 0*valid_mu, color='blue', ls='--', lw=1.5,
         label=r"$x = 0$")
plt.plot(valid_mu, 1 - 1 / valid_mu, color='blue', ls='--', lw=1.5,
         label=r"$x = 1 - \frac{1}{\mu}$")

# Vertical lines
plt.axvline(4, color='black', ls='--', lw=1.8, label=r"$\mu = 4$")
plt.axvline(3.68, color='black', ls='--', lw=1.8, label=r"$\mu \approx 3.68$")

plt.text(3.9, 1.05, "BC", fontsize=32, rotation=0, va='center')
plt.text(3.5, 1.05, "AMC", fontsize=32, rotation=0, va='center')

# Labels and style
plt.xlabel(r"$\mu$")
plt.ylabel(r"$x$")
plt.xlim(2.8, 4.001)
plt.ylim(-0.01, 1.2)
#plt.legend(loc='upper right', frameon=True)
#plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('log_map.eps', format='eps', bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```

```python
# --- Generate bifurcation diagram ---

print("Generating bifurcation diagram...")

# Two bifurcation parameter ranges with different step sizes
bif_params_low = np.arange(3.0, 3.5, 0.008)     # coarse
bif_params_high = np.arange(3.5, 4.1, 0.0008)   # fine

# Concatenate them
bif_params_plot = np.concatenate([bif_params_low, bif_params_high])

timestepss = 600
broadcast_shape = 30

bif_repeated = np.repeat(bif_params_plot, broadcast_shape)
bif_broadcasted = bif_repeated.reshape(len(bif_params_plot), broadcast_shape)

x_list = np.empty((1, timestepss))
x_listfinal = np.empty((len(bif_params_plot), broadcast_shape))

for k, bif_param in enumerate(bif_params_plot):
    rstate0 = rstate_initial.copy()
    for j in range(timestepss):
        r = (1-alpha)*rstate0 + alpha*np.tanh(np.matmul(Lambda, rstate0) + Omega*(bif_param - 0.3))
        rstate0 = r
        x_list[:, j] = np.matmul(W_out, r)

    x_listfinal[k, :] = x_list[:, -broadcast_shape:]
    x_listfinal[k, :] = x_list[:, -broadcast_shape:]

```

```python
A, W_in, W_out, W_b, rstate_initial = model.get_weights()

Lambda = A + np.matmul(W_in, W_out)
Omega = W_b

alpha = 0.86
bif_param = 4.0013                                                            #should be kept as close as possible to the bifurcation point for an accurate estimate of A,B,C in paper

r_sstate = np.zeros(400)
    
x0_list = []
r_list = []
x0 = 0.5                                            #check for any initial condition, you should see similar behaviour
for m in range(300):

    x0 = 3.99*x0*(1 - x0)
    x0_list.append(x0)

x0_array = np.array(x0_list).reshape(300, 1)
warmup_array = x0_array[-41:]  # This will be of shape (41, 1)
    
y_init = warmup_array[-1,:]

for l in range(40):

    rstate_initial1 = ((1- alpha)* r_sstate) + alpha* np.tanh(np.matmul(A, r_sstate) + np.matmul(W_in, warmup_array[l,:].T) + W_b * (bif_param - 0.3))
    r_sstate = rstate_initial1

def h(r_sstate, x):

    r = (1-alpha)*r_sstate + alpha* np.tanh(np.matmul(A, r_sstate) + np.matmul(W_in, x)+ Omega*(bif_param - 0.3))
    
    y = np.matmul(W_out, r)                                                                                                                
                                                                                                                                    
    return y,r

x_array = []
y_array = []


for i in range(50000):
    y1, rstate_initiall = h(r_sstate, y_init)
    x_array.append(y_init)
    y_init = y1
    y_array.append(y1)
    r_list.append(r_sstate)
    r_sstate = rstate_initiall

xx_array = np.array(x_array)
yy_array = np.array(y_array)
y_const = np.full(xx_array.shape, 1)
plt.scatter(x_array, y_array)

#plt.plot(x_array, y_const)
#plt.ylim(0.95,1.02)
plt.show()

#print(x_array[200:300])

xx_flat = np.abs(np.array([x.item() for x in xx_array]))
m = np.argmax(xx_flat)

x_star = x_array[m - 1]
r_star = r_list[m-1]                                                #r[m-1] is correct

lambda_vec = np.matmul(A, r_star) + np.matmul(W_in, x_star)


sech_sq = 1 / (np.cosh(lambda_vec) ** 2)   

product = sech_sq * Omega       #element wise multiplication

result = alpha * np.matmul(W_out, product)

# Print or store result
print("Derivative wrt mu:", result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Flatten arrays
x_array_flat = x_array.flatten()
y_array_flat = np.array(y_array).flatten()

# Keep only positive y-values
mask = (y_array_flat > 0)
x_fit = x_array_flat[mask]
y_fit = y_array_flat[mask]

# Define the fixed constants
d_fixed = 0.48
c_fixed = 0.96575423

# Define the function y = a * (x - d)^n + c
def power_law_shifted(x, a, n):
    return a * np.abs((x - d_fixed))**n + c_fixed

# Range of n values to try
n_values = np.linspace(0.2, 3.0, 50)

best_fit_params = None
best_residuals = np.inf

for n_guess in n_values:
    try:
        # For each fixed n, fit only a
        popt, pcov = curve_fit(
            lambda x, a: power_law_shifted(x, a, n_guess),
            x_fit,
            y_fit,
            p0=[1.0]
        )
        
        # Calculate residuals
        residuals = np.sum(
            (y_fit - power_law_shifted(x_fit, popt[0], n_guess))**2
        )
        
        if residuals < best_residuals:
            best_residuals = residuals
            best_fit_params = (popt[0], n_guess)
            
    except Exception as e:
        continue

if best_fit_params:
    a_fit, n_fit = best_fit_params
    print(f"Best fit: y = {a_fit:.5f} * (x - {d_fixed})^{n_fit:.5f} + {c_fixed:.5f}")
    
    # Plot result
    plt.figure(figsize=(8,5))
    plt.plot(x_fit, y_fit, 'b.', label='Data')
    plt.plot(
        x_fit,
        power_law_shifted(x_fit, a_fit, n_fit),
        'r-', label=f'Fit: a={a_fit:.3f}, n={n_fit:.3f}'
    )
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()
else:
    print("No fit succeeded.")

```
